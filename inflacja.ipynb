```python
import pandas as pd
import matplotlib.pyplot as plt


csv_path = "inflacja.csv"
cpi = pd.read_csv(csv_path, sep=';', decimal=',', encoding='cp1250')

cpi.head()
```

```python
cpi = cpi[cpi['Sposób prezentacji'] == 'Analogiczny miesiąc poprzedniego roku = 100'].copy()
```

```python
cpi.head()
```

```python
cpi.info()
```

```python
cpi = cpi[['Rok','Miesiąc','Wartość']]
```

```python
cpi.info()
```

```python
cpi.columns
```

```python
plt.figure(figsize=(10,8))
# Tworzymy kolumnę daty
# Konwersja na liczby całkowite
cpi['Rok'] = cpi['Rok'].astype(int)
cpi['Miesiąc'] = cpi['Miesiąc'].astype(int)
cpi['Wartość'] = cpi['Wartość'].astype(float)

# Tworzymy kolumnę daty
cpi['date'] = pd.to_datetime(cpi[['Rok','Miesiąc']].assign(day=1))

# Rysujemy wykres liniowy
plt.figure(figsize=(10,5))
plt.plot(cpi['date'], cpi['Wartość'], marker='o', linestyle='-')
plt.title("Wskaźnik CPI - Analogiczny miesiąc poprzedniego roku")
plt.xlabel("Data")
plt.ylabel("Wartość CPI")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
