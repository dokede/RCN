```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

csv_path = "inflacja.csv"
cpi = pd.read_csv(csv_path, sep=';', decimal=',', encoding='cp1250')

cpi.head()
```

```python
cpi = cpi[cpi['Sposób prezentacji'] == 'Analogiczny miesiąc poprzedniego roku = 100'].copy()
```

```python
cpi = cpi[['Rok','Miesiąc','Wartość']]
```

```python
cpi = cpi[cpi['Rok']>2015]
```

```python
cpi['date'] = pd.to_datetime(dict(year=cpi['Rok'], month=cpi['Miesiąc'], day=1))
```

```python
cpi['Wartość'] = cpi['Wartość']-100
```

```python
sns.lineplot(x="date", y="Wartość", data=cpi, marker="o")
plt.show()
```
